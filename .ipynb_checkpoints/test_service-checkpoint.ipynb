```python
example_input_data = [
    {
        "CreditScore": 502,
        "Geography": "France",
        "Gender": "Female",
        "Age": 42,
        "Tenure": 8,
        "Balance": 159660.8,
        "NumOfProducts": 3,
        "HasCrCard": 1,
        "IsActiveMember": 0,
        "EstimatedSalary": 113931.57
    },
    {
        "CreditScore": 699,
        "Geography": "France",
        "Gender": "Female",
        "Age": 39,
        "Tenure": 1,
        "Balance": 0.0,
        "NumOfProducts": 2,
        "HasCrCard": 0,
        "IsActiveMember": 0,
        "EstimatedSalary": 93826.63
    }
]
```

```python
import requests
import json
import numpy as np

# Define the API endpoint URL
public_url = " http://34.55.207.154:3000"
url = f"{public_url}/predict"

request_payload_inner = {"data": example_input_data}
input_data = {"input_data": request_payload_inner}

# The requests library will automatically handle json.dumps() and set Content-Type header
# headers = {'Content-Type': 'application/json'} # No longer strictly needed if using 'json' parameter

print(f"Sending POST request to {url} with input: {input_data}")

try:
    # Send the POST request using the 'json' parameter
    response = requests.post(url, json=input_data)

    # Check if the request was successful
    if response.status_code == 200:
        predictions = response.json()
        print("API Response (predictions):")
        print(predictions)
    else:
        print(f"Error: API returned status code {response.status_code}")
        print(f"Response body: {response.text}")

except requests.exceptions.ConnectionError:
    print("Error: Could not connect to the BentoML service.")
    print("Please ensure 'bentoml serve churn_prediction_service:latest --reload' is running.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
